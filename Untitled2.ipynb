#House Price Prediction

Using an ANN, we are going to predict the prices of real estate properties.

##Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
```

##Load the dataset

```python
dataset = pd.read_csv('data.csv')
dataset.head(n=10)
```

###Understand the shape of dataset

```python
dataset.shape
```

###Describe the dataset

```python
dataset.describe()
```

###Check tensorflow version

```python
tf.__version__
```

###Make matrix of features and dependent variable vector

```python
#Remove the street, country and state zip column from the dataset
```

```python
dataset = dataset.drop(['street','statezip','country'],axis=1)
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,2:].values
y = dataset.iloc[:,1].values
```

```python
print(X)
```

```python
print(y)
```

###Check for missing values

```python
dataset.isnull()
```

```python
print(X[:,12])
```

###Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[12])] , remainder = 'passthrough')
X = (ct.fit_transform(X))
X = X.toarray()
print(X)
```

```python
print(X[1,:])
```

###Split into training and testing set

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

###Feature_Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

##Build ANN

###Initialize ANN

```python
ann = tf.keras.models.Sequential()
```

###Add input layer and first hidden layer

```python
ann.add(tf.keras.layers.Dense(units = 12,activation = 'relu'))
```

###Add second hidden layer

```python
ann.add(tf.keras.layers.Dense(units =12,activation = 'relu'))
```

###Add third hidden layer

```python
ann.add(tf.keras.layers.Dense(units = 12,activation = 'relu'))
```

###Add fourth hidden layer

```python
ann.add(tf.keras.layers.Dense(units = 12 ,activation = 'relu'))
```

###Add the output layer

```python
ann.add(tf.keras.layers.Dense(units = 1 , activation = 'linear'))
```

##Training the ANN

###Compiling the ANN

```python
ann.compile(optimizer='adam',loss='mean_absolute_error')
```

###Training the ANN on the training set

```python
ann.fit(X_train,y_train,batch_size=32,epochs=200)
```

###Predicting the results

```python
y_pred = ann.predict(X_test)
np.set_printoptions(precision = 2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

###Visualizing the results

```python
plt.plot(y_test[0:100],color='red')
plt.plot(y_pred[0:100],color='blue')
plt.show()
```
